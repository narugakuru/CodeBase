```python
#箱线图
samp1 = np.random.normal(loc=0., scale=1., size=100)
samp2 = np.random.normal(loc=1., scale=2., size=100)
samp3 = np.random.normal(loc=0.3, scale=1.2, size=100)
f, ax = plt.subplots(1, 1, figsize=(5,4))
ax.boxplot((samp1, samp2, samp3))
ax.set_xticklabels(['sample 1', 'sample 2','sample 3'])
plt.show()
```

```python
"""从“标准正态”分布返回一个或多个样本。
如果提供了正的int_like参数，randn生成一个形状数组(d0, d1，…)， dn)，
填充了从均值为0和方差为1的单变量“正态”(高斯)分布抽样的随机浮点数。
如果没有提供参数，则返回从分布中随机抽样的单个浮点数。"""
data = np.random.randn(1000, 3)
print(data)
fig, a = plt.subplots(1, 3)
a[0].hist(data, bins=3, histtype="bar")
a[1].hist(data, bins=4, histtype="step", stacked=True)
plt.show()
```

```python
#plt.subplot(‘行’,‘列’,‘编号‘)， projection 投影类型
ax = plt.subplot(111, projection='3d')
#从三个维度构建
z = np.linspace(0, 1, 100)
x = z * np.sin(90 * z)
y = z * np.cos(90 * z)
#调用 ax.plot3D创建三维线图
ax.plot3D(x, y, z)
ax.set_title('3D line plot')
plt.show()
```

```python
from string import ascii_letters
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
columns=list(ascii_letters[26:]))
corr = d.corr()
fig, ax = plt.subplots()
# cmap：色调， cool、 gray； interpolation：边界
ax.imshow(corr, cmap='hot', interpolation='nearest')
ax.set_xticks(np.arange(26))
ax.set_yticks(np.arange(26))
ax.set_xticklabels(list(ascii_letters[26:]))
ax.set_yticklabels(list(ascii_letters[26:]))
plt.show()
```
