```python
# import导入必要的包
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection

#准备数据
df = pd.read_csv('D:/Code/VSCode_Python/data/titanic_data/train.csv')

#特征工程
# 用众数填充Embarked这一列的缺失值
df.Embarked[df.Embarked.isnull()]=df.Embarked.mode()[0]

# 用U填充Cabin中的缺失值,用均值填充Age中的缺失值
df.Cabin.fillna('Unknown',inplace=True)
df.Age.fillna(df.Age.median(),inplace=True)
df.Cabin=df.Cabin.apply(lambda x:x[0])

# 将船票进行计数
tg=df.Ticket.value_counts()

# 将船票号分类(有点问题,待我想想)
df['Ticket_group']=df.Ticket.apply(lambda x:tg[x])

# 找出名字中有价值的信息
df['f_name']=df.Name.apply(lambda x:x[:x.find(',')])
df['title']=df.Name.apply(lambda x:x[x.find(',')+1:x.find('.')])
df['l_name']=df.Name.apply(lambda x:x[x.find('.')+1:])
titles = df['title'].unique()

# 重新根据名字来填充年龄(我觉得这不太应该)
for title in titles:
    m_age = df.Age[df.title==title].median()
    df.loc[(df.Age.isnull()) & (df.title==title),'Age']=m_age
    
# 加入 是否是小孩,这个特征
df.loc[:,'child'] = 0
df.loc[df.Age>=18, 'child'] = 1

# 加入名字长度这个特征
df['name_length'] = df.Name.apply(lambda x: len(x))

# 连续数值型字段的转换
# 增加特征,家人是否获救
df['Family_Survival'] = 0.5
for grp,grp_df in df.groupby(['f_name','Fare']): # 根据名字，船票分组
    if (len(grp_df) != 1):
        for ind,row in grp_df.iterrows():  # 迭代
            smax = grp_df.drop(ind)['Survived'].max()
            smin = grp_df.drop(ind)['Survived'].min()
            passID = row['PassengerId']
            if (smax==1.0):
                df.loc[df['PassengerId'] == passID, 'Family_Survival'] = 1
            elif (smin==0.0):
                df.loc[df['PassengerId'] == passID, 'Family_Survival'] = 0

# 根据船票来修改特征,家人是否获救
for _, grp_df in df.groupby('Ticket'):
    if(len(grp_df) != 1):
        for ind, row in grp_df.iterrows():
            if (row['Family_Survival'] == 0) | (row['Family_Survival']== 0.5):
                smax = grp_df.drop(ind)['Survived'].max()
                smin = grp_df.drop(ind)['Survived'].min()
                passID = row['PassengerId']
                if smax == 1.0:
                    df.loc[df['PassengerId'] == passID, 'Family_Survival'] = 1
                elif smin==0.0:
                    df.loc[df['PassengerId'] == passID, 'Family_Survival'] = 0
                    
# family 增加特征 家庭成员数量
df['family_size'] = df.Parch + df.SibSp + 1
# 三个离散值,方便做独热编码,下同
df['fs_type']='a'
df.loc[(df.family_size>1) & (df.family_size==4),'fs_type'] = 'b'
df.loc[(df.family_size>4),'fs_type'] = 'c'

# # 量化人员 信息
# df['FamilySize'] = df['Parch'] + df['SibSp'] + 1
# df[ 'Family_Single' ] = df[ 'FamilySize' ].map( lambda s : 1 if s == 1 else 0 )
# df[ 'Family_Small' ]  = df[ 'FamilySize' ].map( lambda s : 1 if 2 <= s <= 4 else 0 )
# df[ 'Family_Large' ]  = df[ 'FamilySize' ].map( lambda s : 1 if 5 <= s else 0 )

# Age
df['age_type'] = 'a'
df.loc[(df.Age>16)&(df.Age<=30),'age_type'] = 'b'
df.loc[(df.Age>30)&(df.Age<=40),'age_type'] = 'c'
df.loc[(df.Age)>40,'age_type'] = 'd'

# fare
df['fare_type'] = 'a'
df.loc[(df.Fare>13) & (df.Fare<=25),'fare_type'] = 'b'
df.loc[(df.Fare>25) & (df.Fare<=38),'fare_type'] = 'c'
df.loc[(df.Fare>38) & (df.Fare<=52),'fare_type'] = 'd'
df.loc[(df.Fare>52) & (df.Fare<=102),'fare_type'] = 'e'
df.loc[(df.Fare)>102,'age_type'] = 'e'

# 将一些已经提取出来的特征删除
df=df.drop(['Age','Fare','Parch','SibSp','family_size','f_name','l_name'],axis=1)

#划分特征数据和标签数据
X=df.iloc[:,2:]
y=df.iloc[:,1]
X=X.drop('Name',axis=1)
X=pd.get_dummies(X)
#X=X.select_dtypes(include=np.number)
#X=X.dropna(axis=1)
```
